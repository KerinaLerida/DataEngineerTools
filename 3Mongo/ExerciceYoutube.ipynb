```python
import pandas as pd
import pymongo
```

```python
client = pymongo.MongoClient()
database = client['exercices']
collection = database['youtube']
```

```python
from datetime import datetime
df_youtube = pd.read_csv("./data/USvideos.csv")
df_youtube['date'] = df_youtube['date'].apply(lambda x: datetime.strptime(str(x), '%d.%m').replace(year=2017) if pd.notnull(x) else pd.NaT) # Année 2017 en observant le csv
df_youtube.head(2)
```

## Question 0

### Netoyer les données

```python
df_youtube.info()
```

```python
df_categories = pd.read_json('./data/US_category_id.json')
df_categories = pd.json_normalize(df_categories['items'])
df_categories['id']=df_categories['id'].astype(int)

df_youtube=df_youtube.merge(df_categories[['id','snippet.title']],left_on='category_id',right_on='id')
df_youtube=df_youtube.rename(columns={'snippet.title': 'category'})
df_youtube.head()
```

### Importer les données

```python
data = df_youtube.to_dict(orient='records')
collection.insert_many(data)
```

## Question 1  

```python
videos=collection.find({'channel_title':"Apple"})
for v in videos:
    print(v['title'])
```

## Question 2

```python
nb_catego=collection.aggregate([{"$group": {"_id": "$category", "count": {"$sum":1}}}])

for resultat in nb_catego:
    print(f"Catégorie: {resultat['_id']}, Nombre de documents: {resultat['count']}")
```

## Question 3

```python
documents = collection.find({"tags": {"$exists": True}})

for document in documents:
    tags_list = document["tags"].split("|")
    collection.update_one({"_id": document["_id"]}, {"$set": {"tags": tags_list}})
```

## Question 4

```python
cur=collection.find().sort("views", -1)
for c in cur:
    print(c["channel_title"])
```

## Question 5

```python
nb_catego = collection.aggregate([
    {"$group": {"_id": "$category", "average_views": {"$avg": "$views"}}}
])

for resultat in nb_catego:
    print(f"Catégorie: {resultat['_id']}, Moyenne de vues: {resultat['average_views']}")
```

## Question 6 

```python
resultat = collection.aggregate([
    {"$group": {"_id": "$channel_title", "average_likes": {"$avg": "$likes"}}},
    {"$sort": {"average_likes": -1}},
    {"$limit": 5}
])

for doc in resultat:
    print(f"Chaîne avec la plus grande moyenne de likes: {doc['_id']}, Moyenne de likes: {doc['average_likes']}")
```
